```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (14,6)
```

```python
obs = pd.read_csv("../data/obs_arg.csv", parse_dates=["date"])
obs = obs.set_index("date")
obs = obs.rename(columns={"I":"Is"})
obs
```

```python
obs.plot()
```

```python
obs = obs.rolling(window=7).mean()
```

```python
obs.plot()
```

```python
recovery_rate = (obs.R / obs.Total_Confirmed).tail(5).mean() / 10
recovery_rate
```

```python
mortality_rate = (obs.D / obs.Total_Confirmed).tail(5).mean() / 10
mortality_rate
```

```python
target = obs["2020-04-01":"2020-04-30"].copy()
target = target.drop(columns=["Total_Confirmed"])
target.tail()
```

```python
test = obs["2020-05-01":"2020-05-15"].copy()
test = test.drop(columns=["Total_Confirmed"])
test.head()
```

```python
from SIRD2 import SIRD2
```

```python
model = SIRD2(population=44e6, recovery_rate=recovery_rate, mortality_rate=mortality_rate, alpha=0.2, beta=0.05)
```

```python
model.fit(target, estimate=('beta', "mortality_rate", "recovery_rate"), weights=None, method="Nelder-Mead")
```

```python
y0 = np.concatenate(([500], target.head(1).to_numpy().flatten()))
res = model.simulate(y0, until=42)
```

```python
ax = res.plot()
pd.concat((target, test)).reset_index(drop=True).plot(ax=ax)
```
