```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (14,6)
```

```python
obs = pd.read_csv("../data/obs_arg.csv", parse_dates=["date"])
obs = obs.set_index("date")
obs
```

```python
obs.plot()
```

```python
obs = obs.rolling(window=7, min_periods=1).mean()
```

```python
obs.plot()
```

```python
recovery_rate = (obs.R / obs.Total_Confirmed).tail(5).mean() / 10
recovery_rate
```

```python
mortality_rate = (obs.D / obs.Total_Confirmed).tail(5).mean() / 10
mortality_rate
```

```python
target = obs["2020-04-01":"2020-04-30"].copy()
target = target.drop(columns=["Total_Confirmed"])
target.tail()
```

```python
test = obs["2020-05-01":"2020-05-15"].copy()
test = test.drop(columns=["Total_Confirmed"])
test.head()
```

```python
from SEIR import SIRD
```

```python
model = SIRD(population=44e6, recovery_rate=recovery_rate, mortality_rate=mortality_rate, beta=0.05)
```

```python
x = np.linspace(0, 1, num=30)
y = x**3
y = 1 / y.sum() * y
weights = y.reshape((-1,1)) * 1/3 * np.ones((1,3))
```

```python
weights = np.zeros((30,3))
weights[25:,:] = 1
M = weights[weights.nonzero()].size
weights = 1 / M * weights
```

```python
weights = np.zeros((30,3))
weights[::7,:] = 1
M = weights[weights.nonzero()].size
weights = 1 / M * weights
```

```python
weights.sum()
```

```python
model.fit(target, estimate=('beta', "mortality_rate", "recovery_rate"), weights=None, method="Nelder-Mead")
```

```python
y0 = target.head(1).to_numpy().flatten()
res = model.simulate(y0, until=42)
```

```python
ax = res.plot()
pd.concat((target, test)).reset_index(drop=True).plot(ax=ax)
```
