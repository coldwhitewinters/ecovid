```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
obs = pd.read_csv("../data/obs_arg.csv", parse_dates=["date"])
obs = obs.set_index("date")
obs = obs.drop(columns=["Total_Confirmed"])
obs
```

```python
obs.plot()
```

```python
obs = obs.rolling(window=7, min_periods=1).mean().astype("int64")
```

```python
obs.plot()
```

## SIR

```python
from SEIR import SIR
```

```python
train = []
output = []
params = []

batch_size = 14
for k in range(len(obs) // batch_size):
    train.append(obs.iloc[k*batch_size:(k+1)*batch_size, :])
    
test = train.pop()
```

```python
init_I, init_R, init_D = train[0].iloc[0,:]

model = SIR(population=1e4, 
            init_infected=init_I, 
            init_recovered=init_R, 
            init_dead=init_D, 
           )

model.fit(train[0], 
          estimate=('r_transmission', "r_recovery", "r_mortality", "population"), 
          method="Nelder-Mead",)
          #options={"maxiter": 5000})

res = model.simulate(until=len(train[0]) + 1)
res, y0 = res.iloc[:-1, :], res.iloc[-1, :]
res.index = train[0].index

output.append(res)
params.append(model.params.copy())
```

```python
for k in range(1, len(train)):
    print("----------------------")
    print("Batch", k)
    print()
    print("Setting initial values...")
    model.init_values["infected"] = y0[0]
    model.init_values["recovered"] = y0[1]
    model.init_values["dead"] = y0[2]

    print("Fitting...")
    model.fit(train[k], 
              estimate=('r_transmission', "r_recovery", "r_mortality", "population"), 
              method="Nelder-Mead",) 
              #options={"maxiter": 5000})

    print("Simulating...")
    res = model.simulate(until=len(train[k]) + 1)
    res, y0 = res.iloc[:-1, :], res.iloc[-1, :]
    res.index = train[k].index
    
    output.append(res)
    params.append(model.params.copy())
    print("Done!")
```

```python
model.init_values["infected"] = y0[0]
model.init_values["recovered"] = y0[1]
model.init_values["dead"] = y0[2]

res = model.simulate(until=len(test))
res.index = test.index

output.append(res)
params.append(model.params.copy())
```

```python
ax = pd.concat(output).plot()
obs.plot(ax=ax, style=".")
test.plot(ax=ax, style="--")
```

```python
r_transmission = []
r_recovery = []
r_mortality = []
for d in params:
    r_transmission.append(d["r_transmission"])
    r_recovery.append(d["r_recovery"])
    r_mortality.append(d["r_mortality"])
    
```

```python
plt.plot(r_transmission)
```

```python
plt.plot(r_recovery)
```

```python
plt.plot(r_mortality)
```

```python
R0 = np.array(r_transmission)/np.array(r_recovery)
plt.plot(R0[1:])
```

# SEIR

```python
from SEIR import SEIR
```

```python
train = []
output = []
params = []

batch_size = 14
for k in range(len(obs) // batch_size):
    train.append(obs.iloc[k*batch_size:(k+1)*batch_size, :])
    
test = train.pop()
```

```python
init_I, init_R, init_D = train[0].iloc[0,:]
init_E = 1

model = SEIR(population=1e4, 
            init_infected=init_I, 
            init_recovered=init_R, 
            init_dead=init_D, 
            init_exposed=init_E,
           )

model.fit(train[0], 
          estimate=('r_transmission', "r_recovery", "r_mortality", "r_progression", "population"), 
          method="Nelder-Mead",) #options={"maxiter": 5000})

res = model.simulate(until=len(train[0]) + 1)
res, y0 = res.iloc[:-1, :], res.iloc[-1, :]
res.index = train[0].index

output.append(res)
params.append(model.params.copy())
```

```python
for k in range(1, len(train)):
    print("----------------------")
    print("Batch", k)
    print()
    print("Setting initial values...")
    model.init_values["exposed"] = y0["E"]
    model.init_values["infected"] = y0["I"]
    model.init_values["recovered"] = y0["R"]
    model.init_values["dead"] = y0["D"]

    print("Fitting...")
    model.fit(train[k], 
              estimate=('r_transmission', "r_recovery", "r_mortality", "r_progression", "population"), 
              method="Nelder-Mead",) #options={"maxiter": 5000})

    print("Simulating...")
    res = model.simulate(until=len(train[k]) + 1)
    res, y0 = res.iloc[:-1, :], res.iloc[-1, :]
    res.index = train[k].index
    
    output.append(res)
    params.append(model.params.copy())
    print("Done!")
```

```python
model.init_values["exposed"] = y0["E"]
model.init_values["infected"] = y0["I"]
model.init_values["recovered"] = y0["R"]
model.init_values["dead"] = y0["D"]

res = model.simulate(until=len(test))
res.index = test.index

output.append(res)
params.append(model.params.copy())
```

```python
ax = pd.concat(output)[["E", "I", "R", "D"]].plot()
obs.plot(ax=ax, style=".")
test.plot(ax=ax, style="+")
```
